```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
```

```python
csvdata=pd.read_csv("Medical Expenses of Smoking.csv")
csvdata
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
 
# y-axis in bold
rc('font', weight='bold')
 
# Values of each group
bars1 = csvdata['Direct Cost-Medical Expenses']
bars2 = csvdata['Indirect Cost-Related to Illness']
bars3 = csvdata['Indirect Cost-Related to Death']
 
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
 
# The position of the bars on the x-axis
r = csvdata['Year']
 
# Bar width and Figure size
barWidth = 0.5
plt.figure(figsize=(12,6))

# Create green bars
plt.bar(r, bars1, color='#60c5ba', edgecolor='white', width=barWidth,label='Direct Cost-Medical Expenses')
# Create yellow bars (middle), on top of the firs ones
plt.bar(r, bars2, bottom=bars1, color='#feee7d', edgecolor='white', width=barWidth,label='Indirect Cost-Related to Illness')
# Create red bars (top)
plt.bar(r, bars3, bottom=bars, color='#9B90C2', edgecolor='white', width=barWidth,label='Indirect Cost-Related to Death')
plt.legend() 

# Custom X axis, Y axis and title
plt.xticks(r, fontweight='bold')
plt.xlabel("Year",fontweight='bold',fontsize=20)
plt.ylabel("100 Million NTD",fontweight='bold',fontsize=20)
plt.title("Direct and Indirect Cost Caused by Smoking",fontweight='bold',fontsize=30)

 
# Show graphic
plt.show()
```

```python
from matplotlib.colors import ListedColormap
df=pd.read_csv("medical expenses.csv")
df.set_index('Cost')\
  .reindex(df.set_index('Cost').sum().sort_values().index, axis=1)\
  .T.plot(kind='bar', stacked=True,
          colormap=ListedColormap(sns.color_palette("Set3", 10)), 
          figsize=(12,6))\
  .set(xlabel='Year',ylabel='100 Million',title='Direct and Indirect Cost Caused by Smoking')
```
